Vamos a tratar el Knapsack 0−1 Problem o Problema de la mochila binario, con modelo matemático:
$$ Max V = \sum_{j=i}^{n} c_jx_j = c_1x_1 + c_2x_2 + ... + c_nx_n $$ 
$$ s.a. \longrightarrow \sum_{j=i}^{n} a_jx_j \leq b \mid x \in (0,1); j = 1,...,n $$

Vamos a generar 50 productos, con un pesos aleatorios entre cinco gramos (150g) y diez kilogramos (10kg→10.000g), y costo entre mil pesos (COP 1000) y un millon de pesos (COP 1'000.000). Para esto nos ayudaremos de la libreria de Python Numpy. 

```python
import numpy as np 

products = {}
n = 50

for i in range(n):
    products['P{}'.format(i+1)] = {
        'weight': np.random.randint(150, 10000),
        'cost': np.random.randint(1000, 1000000),
    }
    
products
```

El volumen de la mochila será dos tercios de la suma total de todos los pesos de los productos

$$ Vol = \frac{2}{3} \sum_{i=1}^{n=50} weight_i $$

```python
backpack_volume = round(2/3*sum(map(lambda product: product['weight'], products.values())))
print('Backpack volume:', backpack_volume)
```

Vamos a utilizar como **Indicador de sensiblidad (I.S.)**, la división entre el costo y peso de cada producto. Ya que nos permite llevar los productos que mejor razón entre costo y peso, es decir, nos llevaremos los productos que por unidad de peso, cuestan mas. 

```python
for name, value in products.items():
    products[name]['c/w'] = round(products[name]['cost']/products[name]['weight'],2)
    
products
```

A partir del anterior expuesto I.S. vamos a generar 2 esquemas, colocando los productos que tengan mayor razón entre costo y peso **(c/w)**, primero vamos a ordenar los productos de mayor a menor de acuerdo al c/w

```python
ordered_products = dict(sorted(products.items(), key=lambda product: product[1]['c/w'], reverse=True))

ordered_products
```

Construimos una función que nos devuelve la evaluación de un esquema, la evaluación nos arroja una tupla de 3 posiciones en la cual se encuentra sumatoria del peso, costo de los productos y el factor de infactibilidad del esquema.

Tambien se construye una función que nos imprime los productos que contiene un esquema, con su respectivo resultado de la función objetivo y el volumen.

```python
def evaluation_schema(schema):
    weight = 0
    cost = 0
    infeasibility_factor = 0
    for name, value in schema.items():
        if value:
            weight += products[name]['weight']
            cost += products[name]['cost']
    return (weight, cost, max(0, weight-backpack_volume))

def print_schema(schema):
    print('----------------------')
    print('This backpack contains:')
    for name, value in schema.items():
        if value:
            print(name, end=' ')
    evaluation = evaluation_schema(schema)
    print('\nVolume {}\nO.B {} \nInfeasibility Factor {}'.format(evaluation[0], evaluation[1], evaluation[2]))
    print('----------------------')
    
```

Para construir los dos esquemas de buena calidad, lo que utilizaremos, para el primer esquema es agregar el 10% del total de productos que mayor razón de costo sobre peso (c/w), osea, los primeros 5 productos que mayor tienen c/w. Y para el segundo el 5% del total de los productos

```python
first_schema = {}
for name, value in products.items():
    first_schema[name] = 0 

print('First 5 products')
for name, value in list(ordered_products.items())[0:5]:
    print(name)
    first_schema[name] = 1 
    
print_schema(first_schema)
```

```python
second_schema = {}
for name, value in products.items():
    second_schema[name] = 0 

print('First 3 products')
for name, value in list(ordered_products.items())[0:3]:
    print(name)
    second_schema[name] = 1 
    
print_schema(second_schema)
```

Lo primero que vamos a hacer, es construir la lista restringida de candidatos **(LRC)** para ambos esquemas

```python
print('LRC of first schema')
candidate_products_first_schema = ordered_products.copy()

for name, value in first_schema.items():
    if value:
        del(candidate_products_first_schema[name])
        
candidate_products_first_schema
```

```python
print('LRC of second schema')
candidate_products_second_schema = ordered_products.copy()

for name, value in second_schema.items():
    if value:
        del(candidate_products_second_schema[name])
        
candidate_products_second_schema
```

Ahora, lo que haremos es construir los vecinos de nuestros dos esquemas.

```python
neighbours_first_schema = []

number_of_neighbors_first_schema = np.random.randint(2, 6)

print('{} neighbours in first schema'.format(number_of_neighbors_first_schema))
for i in range(number_of_neighbors_first_schema):
    neighbor = first_schema.copy()
    product_to_add = list(candidate_products_first_schema.keys())[i]
    neighbor[product_to_add] = 1
    evaluation = evaluation_schema(neighbor)
    neighbours_first_schema.append({'neighbor': neighbor, 'volume': evaluation[0],
                                    'objetive_function': evaluation[1],
                                    'infeasibility_factor': evaluation[2]})
    
for index, neighbor in enumerate(neighbours_first_schema):
    print('\nNeighbor {}'.format(index+1))
    print_schema(neighbor['neighbor'])
```

```python
neighbours_second_schema = []

number_of_neighbors_second_schema = np.random.randint(2, 6)

print('{} neighbours in second schema'.format(number_of_neighbors_second_schema))
for i in range(number_of_neighbors_second_schema):
    neighbor = first_schema.copy()
    product_to_add = list(candidate_products_second_schema.keys())[i]
    neighbor[product_to_add] = 1
    evaluation = evaluation_schema(neighbor)
    neighbours_second_schema.append({'neighbor': neighbor, 'volume': evaluation[0],
                                    'objetive_function': evaluation[1],
                                    'infeasibility_factor': evaluation[2]})
    
for index, neighbor in enumerate(neighbours_second_schema):
    print('\nNeighbor {}'.format(index+1))
    print_schema(neighbor['neighbor'])
```

De los vecinos obtenidos, vamos a escoger un vecino de forma al azar de forma **equiprobable**

```python
chosen_schema = np.random.randint(1, len(neighbours_first_schema)+1)

print('Schema chosen at random for the first schema was {}'.format(chosen_schema))

first_schema = neighbours_first_schema[chosen_schema-1]['neighbor']

print_schema(first_schema)
```

```python
chosen_schema = np.random.randint(1, len(neighbours_second_schema)+1)

print('Schema chosen at random for the second schema was {}'.format(chosen_schema))

second_schema = neighbours_second_schema[chosen_schema-1]['neighbor']

print_schema(second_schema)
```
