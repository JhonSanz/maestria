Vamos a tratar el Knapsack 0−1 Problem o Problema de la mochila binario, con modelo matemático:
$$ Max V = \sum_{j=i}^{n} c_jx_j = c_1x_1 + c_2x_2 + ... + c_nx_n $$ 
$$ s.a. \longrightarrow \sum_{j=i}^{n} a_jx_j \leq b \mid x \in (0,1); j = 1,...,n $$

Vamos a generar 50 productos, con un pesos aleatorios entre cinco gramos (150g) y diez kilogramos (10kg→10.000g), y costo entre mil pesos (COP 1000) y un millon de pesos (COP 1'000.000). Para esto nos ayudaremos de la libreria de Python Numpy. 

```python
import numpy as np 

products = []

for i in range(50):
    product = {
        'name': 'P{}'.format(i+1),
        'weight': np.random.randint(150, 10000),
        'cost': np.random.randint(1000, 1000000),
    }
    products.append(product)
    
products
```

El volumen de la mochila será dos tercios de la suma total de todos los pesos de los productos

$$ Vol = \frac{2}{3} \sum_{i=1}^{n=50} weight_i $$

```python
backpack_volume = round(2*sum(map(lambda product: product['weight'], products))/3,2)
print('Backpack volume: ', backpack_volume)
```

Vamos a utilizar como **Indicador de sensiblidad (I.S.)**, la división entre el costo y peso de cada producto. Ya que nos permite llevar los productos que mejor razón entre costo y peso, es decir, nos llevaremos los productos que por unidad de peso, cuestan mas. 

```python
for product in products:
    product['c/w'] = round(product['cost']/product['weight'],2)
    
products
```

A partir del anterior expuesto I.S. vamos a generar 2 esquemas, colocando los productos que tengan mayor razón entre costo y peso **(c/w)**, primero vamos a ordenar los productos de mayor a menor de acuerdo al c/w

```python
ordered_products = sorted(products, key=lambda product: product['c/w'], reverse=True)

ordered_products
```

Construimos una función que nos devuelve la evaluación de un esquema, la evaluación nos arroja una dupla de la sumatoria del peso y costo de los productos del esquema.

```python
def evaluation_schema(schema):
    weight = 0
    cost = 0
    for index in range(len(schema)):
        if schema[index]:
            weight += products[index]['weight']
            cost += products[index]['cost']
    return (weight, cost)
```

Para construir el primer esquema, lo que utilizaremos, es agregar los productos que mayor razón de costo sobre peso (c/w) hasta que, cuando se le vaya a agregar el siguiente producto en la lista ordenada de productos, la sumatoria del peso de los productos del esquema este por encima del volumen de la mochila.

```python
first_schema = [0]*50

for product in ordered_products:
    product_index = int(product['name'].replace('P', '')) - 1
    schema_aux = first_schema.copy()
    schema_aux[product_index] = 1
    evaluation = evaluation_schema(schema_aux)
    if evaluation[0] < backpack_volume: 
        first_schema = schema_aux
    else: break
    
print(first_schema, evaluation_schema(first_schema))
```

Para el segundo esquema, utilizaremos la misma practica anteriormente, con la diferencia de que, se seguira recorriendo todos los productos organizados, y evaluando si hay otro producto que pueda entrar en el esquema y se cumpla con la solución del problema.

```python
second_schema = [0]*50

for product in ordered_products:
    product_index = int(product['name'].replace('P', '')) - 1
    schema_aux = second_schema.copy()
    schema_aux[product_index] = 1
    evaluation = evaluation_schema(schema_aux)
    if evaluation[0] < backpack_volume: 
        second_schema = schema_aux
    
    
print(second_schema, evaluation_schema(second_schema))
```
