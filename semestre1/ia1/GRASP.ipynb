Vamos a tratar el Knapsack 0−1 Problem o Problema de la mochila binario, con modelo matemático:
$$ Max V = \sum_{j=i}^{n} c_jx_j = c_1x_1 + c_2x_2 + ... + c_nx_n $$ 
$$ s.a. \longrightarrow \sum_{j=i}^{n} a_jx_j \leq b \mid x \in (0,1); j = 1,...,n $$

Vamos a generar 50 productos, con un pesos aleatorios entre cinco gramos (150g) y diez kilogramos (10kg→10.000g), y costo entre mil pesos (COP 1000) y un millon de pesos (COP 1'000.000). Para esto nos ayudaremos de la libreria de Python Numpy. 

````python
import numpy as np 

products = {}
n = 50

for i in range(n):
    products['P{}'.format(i+1)] = {
        'weight': np.random.randint(150, 10000),
        'cost': np.random.randint(1000, 1000000),
    }
    
products
````

El volumen de la mochila será dos tercios de la suma total de todos los pesos de los productos

$$ Vol = \frac{2}{3} \sum_{i=1}^{n=50} weight_i $$

````python
backpack_volume = round(2/3*sum(map(lambda product: product['weight'], products.values())))
print('Backpack volume:', backpack_volume)
````

Vamos a utilizar como **Indicador de sensiblidad (I.S.)**, la división entre el costo y peso de cada producto. Ya que nos permite llevar los productos que mejor razón entre costo y peso, es decir, nos llevaremos los productos que por unidad de peso, cuestan mas. 

````python
for name, value in products.items():
    products[name]['c/w'] = round(products[name]['cost']/products[name]['weight'],2)
    
products
````

A partir del anterior expuesto I.S. vamos a generar 2 esquemas, colocando los productos que tengan mayor razón entre costo y peso **(c/w)**, primero vamos a ordenar los productos de mayor a menor de acuerdo al c/w

````python
ordered_products = dict(sorted(products.items(), key=lambda product: product[1]['c/w'], reverse=True))

ordered_products
````

Construimos una función que nos devuelve la evaluación de un esquema, la evaluación nos arroja una tupla de 3 posiciones en la cual se encuentra sumatoria del peso, costo de los productos y el factor de infactibilidad del esquema.

Tambien se construye una función que nos imprime los productos que contiene un esquema, con su respectivo resultado de la función objetivo, el volumen y el indicador sensibilidad de costo sobre volumen.

````python
def evaluation_schema(schema):
    weight = 0
    cost = 0
    infeasibility_factor = 0
    for name, value in schema.items():
        if value:
            weight += products[name]['weight']
            cost += products[name]['cost']
    return (weight, cost, max(0, weight-backpack_volume), round(cost/weight, 2))

def print_schema(schema):
    print('----------------------')
    print('This backpack contains:')
    for name, value in schema.items():
        if value:
            print(name, end=' ')
    evaluation = evaluation_schema(schema)
    print('\nVolume {}\nO.B {} \nInfeasibility Factor {} \nc/w {}'.format(evaluation[0], evaluation[1], 
                                                                       evaluation[2], evaluation[3]))
    print('----------------------')
    
````

Para construir los dos esquemas de buena calidad, lo que utilizaremos, para el primer esquema es agregar el 10% del total de productos que mayor razón de costo sobre peso (c/w), osea, los primeros 5 productos que mayor tienen c/w. Y para el segundo el 5% del total de los productos

````python
first_schema = {}
for name, value in products.items():
    first_schema[name] = 0 

print('First 5 products')
for name, value in list(ordered_products.items())[0:5]:
    print(name)
    first_schema[name] = 1 
    
print_schema(first_schema)
````

````python
second_schema = {}
for name, value in products.items():
    second_schema[name] = 0 

print('First 3 products')
for name, value in list(ordered_products.items())[0:3]:
    print(name)
    second_schema[name] = 1 
    
print_schema(second_schema)
````

Lo primero que vamos a hacer, es construir la lista restringida de candidatos **(LRC)** para ambos esquemas

````python
print('LRC of first schema')
candidate_products_first_schema = ordered_products.copy()

for name, value in first_schema.items():
    if value:
        del(candidate_products_first_schema[name])
        
candidate_products_first_schema
````

````python
print('LRC of second schema')
candidate_products_second_schema = ordered_products.copy()

for name, value in second_schema.items():
    if value:
        del(candidate_products_second_schema[name])
        
candidate_products_second_schema
````

Ahora, lo que haremos es construir los vecinos de nuestros dos esquemas.

````python
neighbours_first_schema = []

number_of_neighbors_first_schema = np.random.randint(2, 6)

print('{} neighbours in first schema'.format(number_of_neighbors_first_schema))
for i in range(number_of_neighbors_first_schema):
    neighbor = first_schema.copy()
    product_to_add = list(candidate_products_first_schema.keys())[i]
    neighbor[product_to_add] = 1
    evaluation = evaluation_schema(neighbor)
    neighbours_first_schema.append(neighbor)
    
for index, neighbor in enumerate(neighbours_first_schema):
    print('\nNeighbor {}'.format(index+1))
    print_schema(neighbor)
````

````python
neighbours_second_schema = []

number_of_neighbors_second_schema = np.random.randint(2, 6)

print('{} neighbours in second schema'.format(number_of_neighbors_second_schema))
for i in range(number_of_neighbors_second_schema):
    neighbor = first_schema.copy()
    product_to_add = list(candidate_products_second_schema.keys())[i]
    neighbor[product_to_add] = 1
    evaluation = evaluation_schema(neighbor)
    neighbours_second_schema.append(neighbor)
    
for index, neighbor in enumerate(neighbours_second_schema):
    print('\nNeighbor {}'.format(index+1))
    print_schema(neighbor)
````

De los vecinos obtenidos, vamos a escoger un vecino de forma al azar de forma **equiprobable**, y el esquema actual lo guardamos en una lista de respuestas llamada **Incumbentes**

````python
chosen_schema = np.random.randint(1, len(neighbours_first_schema)+1)

incumbents_first_schema = [first_schema]

print('Schema chosen at random for the first schema was {}'.format(chosen_schema))

first_schema = neighbours_first_schema[chosen_schema-1]

print_schema(first_schema)
````

````python
chosen_schema = np.random.randint(1, len(neighbours_second_schema)+1)

incumbents_second_schema = [second_schema]

print('Schema chosen at random for the second schema was {}'.format(chosen_schema))

second_schema = neighbours_second_schema[chosen_schema-1]

print_schema(second_schema)
````

Como se logra ver el esquema escogido tiene un factor de infactibilidad de cero, entonces volveremos a hacer el anterior proceso de calcular cierta cantidad aleaotoria de vecinos, escoger uno de ellos de forma aleatoria hasta que el factor de infactibilidad sea diferente de cero. Esto anterior con nuestros dos esquemas.

````python
iteration = 2

while True:
    print('------------------------------')
    print('Iteration {}'.format(iteration))
    print('LRC of first schema')
    candidate_products_first_schema = ordered_products.copy()
    for name, value in first_schema.items():
        if value:
            del(candidate_products_first_schema[name])
    
    neighbours_first_schema = []
    number_of_neighbors_first_schema = np.random.randint(2, 6)

    print('{} neighbours in first schema'.format(number_of_neighbors_first_schema))
    for i in range(number_of_neighbors_first_schema):
        neighbor = first_schema.copy()
        product_to_add = list(candidate_products_first_schema.keys())[i]
        neighbor[product_to_add] = 1
        evaluation = evaluation_schema(neighbor)
        neighbours_first_schema.append(neighbor)

    for index, neighbor in enumerate(neighbours_first_schema):
        print('\nNeighbor {}'.format(index+1))
        print_schema(neighbor)
        
    incumbents_first_schema.append(first_schema)

    chosen_schema = np.random.randint(1, len(neighbours_first_schema)+1)
    
    print('\nSchema chosen at random for the first schema was {}'.format(chosen_schema))
    first_schema = neighbours_first_schema[chosen_schema-1]
    print_schema(first_schema)
    if evaluation_schema(first_schema)[2] > 0: break
    iteration += 1

    
````

````python
iteration = 2

while True:
    print('------------------------------')
    print('Iteration {}'.format(iteration))
    print('LRC of second schema')
    candidate_products_second_schema = ordered_products.copy()
    for name, value in second_schema.items():
        if value:
            del(candidate_products_second_schema[name])
    
    neighbours_second_schema = []
    number_of_neighbors_second_schema = np.random.randint(2, 6)

    print('{} neighbours in second schema'.format(number_of_neighbors_second_schema))
    for i in range(number_of_neighbors_second_schema):
        neighbor = second_schema.copy()
        product_to_add = list(candidate_products_second_schema.keys())[i]
        neighbor[product_to_add] = 1
        evaluation = evaluation_schema(neighbor)
        neighbours_second_schema.append(neighbor)

    for index, neighbor in enumerate(neighbours_second_schema):
        print('\nNeighbor {}'.format(index+1))
        print_schema(neighbor)
    
    incumbents_second_schema.append(second_schema)
        
    chosen_schema = np.random.randint(1, len(neighbours_second_schema)+1)
    
    print('\nSchema chosen at random for the second schema was {}'.format(chosen_schema))
    second_schema = neighbours_second_schema[chosen_schema-1]
    print_schema(second_schema)
    if evaluation_schema(second_schema)[2] > 0: break
    iteration += 1
````

Ahora que el factor de factibilidad es mayor que 0, vamos a guardar el anterior esquema en una lista de soluciones, que despues vamos a comparar con otros posibles soluciones y encontrar la mejor. Consiguiente, debemos hacer un proceso llamado **Path Relinking** 

````python
possible_solutions_first_schema = [incumbents_first_schema[-1]]

incumbents_first_schema.append(first_schema)

print('List of possible solutions of first schema\n')
for index, schema in enumerate(possible_solutions_first_schema):
    print('S{}'.format(index+1))
    print_schema(schema)

print('\n----------------------')
print('List of incumbents of first schema')
for index, schema in enumerate(incumbents_first_schema):
    print('\nX{}'.format(index+1))
    print_schema(schema)
````

````python
possible_solutions_second_schema = [incumbents_second_schema[-1]]

print('List of possible solutions of second schema\n')
for index, schema in enumerate(possible_solutions_second_schema):
    print('S{}'.format(index+1))
    print_schema(schema)
    
incumbents_second_schema.append(second_schema)

print('\n----------------------')
print('List of incumbents of second schema')
for index, schema in enumerate(incumbents_second_schema):
    print('\nX{}'.format(index+1))
    print_schema(schema)
````

Vamos a codificar una función que nos devuelve la diferencia simetrica entre dos esquemas, que la utilizaremos en el Path Relinking.

La diferencia simetrica, nos devuelve un numero que significa la cantidad de articulos que estan en el esquema a y esquema b, a excepcion de los elementos que estan en ambos.

````python
def symmetric_difference(schema_a, schema_b):
    result = {}
    keys = schema_a.keys()
    for key in keys:
        result[key] = 1 if schema_a[key] != schema_b[key] else 0
    return result
````

Ahora calculamos la diferencia simetrica entre todos los incumbentes, incluido el incumbente infactible.

````python
print('Symmetric differences of incumbents of first schema')

symmetric_differences_first_schema = []

for index, schema_a in enumerate(incumbents_first_schema):
    for j, schema_b in enumerate(incumbents_first_schema[index+1:-1]):
        result_schema = symmetric_difference(schema_a, schema_b)
        print('\nX{} Δ X{} = {}'.format(index+1, index+1+j+1, sum(result_schema.values())))
        print_schema(result_schema)
        symmetric_differences_first_schema.append({'result': result_schema, 
                                                   'index_a': index+1, 'index_b': index+1+j+1})
````

````python
print('Symmetric differences of incumbents of second schema')

symmetric_differences_second_schema = []

for index, schema_a in enumerate(incumbents_second_schema):
    for j, schema_b in enumerate(incumbents_second_schema[index+1:-1]):
        result_schema = symmetric_difference(schema_a, schema_b)
        print('\nX{} Δ X{} = {}'.format(index+1, index+1+j+1, sum(result_schema.values())))
        print_schema(result_schema)
        symmetric_differences_second_schema.append({'result': result_schema, 
                                                   'index_a': index+1, 'index_b': index+1+j+1})
````

Ahora despues de calcular las diferencias simetricas, vamos a construir la lista elite.

Para escoger los esquemas que van a ir en la lista de elite, lo que vamos a hacer como indicador de sensibilidad es multiplicar la cantidad de articulos que tenga la mochila por el costo sobre peso.

````python
print('I.S to choose elite list of first schema\n')

for schema in symmetric_differences_first_schema:
    evaluation = evaluation_schema(schema['result'])
    schema['IS'] = round(evaluation[3] * sum(schema['result'].values()),2)
    print('I.S. of X{} Δ X{} = {}'.format(schema['index_a'], schema['index_b'], schema['IS']))
````

````python
print('I.S to choose elite list of second schema\n')

for schema in symmetric_differences_second_schema:
    evaluation = evaluation_schema(schema['result'])
    schema['IS'] = round(evaluation[3] * sum(schema['result'].values()),2)
    print('I.S. of X{} Δ X{} = {}'.format(schema['index_a'], schema['index_b'], schema['IS']))
````

Ahora, vamos a ordenar de mayor a menor y escoger el 20% de la cantidad de productos que son 50. Entonces escogeremos los primeros 8.

````python
print('Elite list of first schema')
elite_list_first_schema = list(sorted(symmetric_differences_first_schema, 
                                      key=lambda schema: schema['IS'], reverse=True))[0:8]

for schema in elite_list_first_schema:
    print('\nX{} Δ X{} with IS = {}'.format(schema['index_a'], schema['index_b'], schema['IS']))
    print_schema(schema['result'])
````

````python
print('Elite list of second schema')
elite_list_second_schema = list(sorted(symmetric_differences_second_schema, 
                                       key=lambda schema: schema['IS'], reverse=True))[0:8]

for schema in elite_list_second_schema:
    print('\nX{} Δ X{} with IS = {}'.format(schema['index_a'], schema['index_b'], schema['IS']))
    print_schema(schema['result'])
````

Ahora, vamos a volver a hacer diferencia simetrica de la lista elite con la solución guia, que es la mejor incumbente hallada en la iteración.

````python
print('SG - Guide solution')
print_schema(possible_solutions_first_schema[-1])

schemes_to_choose_first_schema = []

print('\nList of schemes to choose')
for index, schema in enumerate(elite_list_first_schema):
    result_schema = symmetric_difference(possible_solutions_first_schema[-1], schema['result'])
    print('\nPOS = {} \nSG Δ (X{} Δ X{}) = {}'.format(index+1, 
                                                      schema['index_a'], schema['index_b'], 
                                                      sum(result_schema.values())))
    print_schema(result_schema)
    schemes_to_choose_first_schema.append({'schema': result_schema,
                                           'name': 'SG Δ (X{} Δ X{})'.format(
                                               schema['index_a'], 
                                               schema['index_b'], sum(result_schema.values()))})
````

````python
print('SG - Guide solution')
print_schema(possible_solutions_second_schema[-1])

schemes_to_choose_second_schema = []

print('\nList of schemes to choose')
for index, schema in enumerate(elite_list_second_schema):
    result_schema = symmetric_difference(possible_solutions_second_schema[-1], schema['result'])
    print('\nPOS = {} \nSG Δ (X{} Δ X{}) = {}'.format(index+1,
                                                      schema['index_a'], schema['index_b'], 
                                                      sum(result_schema.values())))
    print_schema(result_schema)
    schemes_to_choose_second_schema.append({'schema': result_schema,
                                           'name': 'SG Δ (X{} Δ X{})'.format(
                                               schema['index_a'], 
                                               schema['index_b'], sum(result_schema.values()))})
````

Ahora, escogeremos un esquema de la lista de esquemas que obtuvimos, la forma en que lo escogeremos será de forma equiprobable.

````python
chosen_schema = np.random.randint(1, len(schemes_to_choose_first_schema)+1)

cousin_schema_first_schema = schemes_to_choose_first_schema[chosen_schema]

print('\nSchema chosen at random for the cousin schema of first schema was {}'.format(
    cousin_schema_first_schema['name']))
print_schema(cousin_schema_first_schema['schema'])
````

````python
chosen_schema = np.random.randint(1, len(schemes_to_choose_second_schema)+1)

cousin_schema_second_schema = schemes_to_choose_second_schema[chosen_schema]

print('\nSchema chosen at random for the cousin schema of second schema was {}'.format(
    cousin_schema_second_schema['name']))
print_schema(cousin_schema_second_schema['schema'])
````
