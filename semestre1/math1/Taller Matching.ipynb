Según las definiciones:

1. Un matching "es un subconjunto de arcos independientes en G, tal que ningún par de arcos de M tienen vértices en común.". Entonces primero que todo podemos asegurar que efectivamente los arcos señalados en el grafo de la figura 1 conforman un matching; siendo este descrito por los arcos:
    - (g, c)
    - (a, b)
    - (h, d)
    - (i, e)
    - (f, j)

2. Un "match es maximum si no existe otro match M' tal que |M'| > |M|". Es decir que la carninalidad del conjunto de arcos del match sea la mayor posible entre todos los match que puedan conformarse en el grafo. Entonces, si comparamos la figura 1 con la figura 2, podemos observar primero que todo, que la figura 2 efectivamente conforma un match, descrito por los arcos:
    - (c, a)
    - (b, f)
    - (d, e)
    - (h, i)
    - (k, g)
    - (l, j)

cuya carninalidad es |M'| = 6, mientras que |M| = 5; entonces el match de la figura 1 no puede ser maximum.

3. Un "match M es maximal si no existe otro match M' que contenga estrictamente a M. Como podemos observar en la figura 1, si quisiéramos crear un match que contenga a M sería necesario construir un nuevo match M' agregando arcos a M, pero según la definición del punto 1, agregando un arco nuevo a M estaríamos violando la definción de match, ya que habrían arcos con nodos repetidos dentro del conjunto M.

Por lo tanto podemos decir que el match de la figura 1 M es solamente maximal.

##### 3. 
Un problema de conteo. ¿Cuantos matchs perfectos hay en $ K_4 $? ¿ En $ K_6 $ ? Y en general, ¿En $ K_{2n} $ ? 

##### Desarrollo.

Para $ K_4 \rightarrow n = 2 = K_{2*2} $

Tenemos 4 vertices que necesitamos emparejar. Vamos a tratar este problema como un problema de conjuntos y combinatoria.

Según la teoria de conjuntos no ordenados tenemos que un conjunto de n elementos se puede dividir en k subconjuntos no ordenados de r elementos cada uno de la siguiente manera $$ \frac{m!}{k!*(r!)^k} $$

Entonces llevando esto a nuestro problema tenemos que, $ m = |V| $ que son la cantidad de vertices, $ r = 2 $ ya que son parejas osea conjuntos de 2 elementos y $ k = |V|/2 $ que son el numero de parejas a conseguir.

$$ m = 2n = 2*2 = 4 $$
$$ k = \frac{4}{2} = 2 $$
$$ r = 2 $$

$$ \frac{4!}{2!*(2!)^2} = \frac{24}{2*4} = \frac{24}{8} = 3 $$


Para $ K_6 \rightarrow n = 3 = K_{2*3} $

$$ m = 2n = 2*3 = 6 $$
$$ k = \frac{6}{2} = 3 $$
$$ r = 2 $$

$$ \frac{6!}{3!*(2!)^3} = \frac{720}{6*8} = \frac{720}{48} = 15 $$

##### 6. 
Un cuadrado latino es un arreglo n × n en el que cada fila y cada columna contienen los números de 1 a n en algún orden. Un rectángulo latino es un cuadrado latino incompleto.
Figura 10 a) Un cuadrado latino. b) Un rectángulo latino.
Demuestre que cualquier rectángulo latino siempre se puede extender hasta tener un cuadrado latino.

![png](images/match-ej-6.png)

##### Desarrollo.

La definición de un rectangulo latino es un $ r×n $ matriz (donde $ r \leq n $), usando n símbolos, por lo general los números $ (1, 2, 3, ..., n) $ o $ (0, 1, ..., n-1) $ como sus entradas, sin que ningún número aparezca más de una vez en cualquier fila o columna. 

Y un rectangulo latino lo podemos extender hasta un cuadrado latino, de manear que vamos a agregar $ n-r $ filas, cumpliendo la condición de que ningun número puede aparecer en cualquier fila o columna
